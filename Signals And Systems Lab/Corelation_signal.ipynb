```python
import numpy as np
import matplotlib.pyplot as plt

# genarat  singal
fs=100
t=np.linspace(0,1,fs,endpoint=False)
freq=5
ferq1=10
# sine
sing=np.sin(2*np.pi*freq*t)

cos=np.cos(2*np.pi*ferq1*t)

autocoreletion=np.correlate(sing,sing,mode='full')
cros_corelation=np.correlate(sing,cos,mode='full')
lags = np.arange(-len(sing) +1, len(sing))

plt.figure(figsize=(10,12))

plt.subplot(4,1,1)
plt.plot(sing)
plt.xlabel("legs")
plt.ylabel("Amplitude")
plt.subplot(4,1,2)
plt.plot(cos)
plt.xlabel("legs")
plt.ylabel("Amplitude")
plt.subplot(4,1,3)
plt.plot(lags,autocoreletion)
plt.xlabel("legs")
plt.ylabel("Amplitude")
plt.subplot(4,1,4)
plt.plot(cros_corelation)
plt.xlabel("legs")
plt.ylabel("Amplitude")

plt.tight_layout
plt.show()

```
